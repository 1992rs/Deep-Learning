```python
import pandas as pd
df = pd.read_csv('Imbalanced_data.csv') 
df.head()
```

```python
df.shape
```

```python
df.columns = ['X_column' , 'Y_column', 'Outcome']
```

```python
df
```

```python
df.value_counts()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
print(X)
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
X['X_column'] = lb.fit_transform(X['X_column'])
X['Y_column'] = lb.fit_transform(X['Y_column'])
print(X)
```

```python
from sklearn.model_selection import train_test_split 
X_train , X_test ,y_train ,y_test = train_test_split(X , y , test_size = 0.3 , random_state = 10) 
```

```python
from sklearn.neighbors import KNeighborsClassifier 
model = KNeighborsClassifier() 
model.fit(X_train , y_train) 
y_pred = model.predict(X_test) 
```

```python
from sklearn.metrics import accuracy_score 
print(accuracy_score(y_test, y_pred))
```

```python
pd.crosstab(y_test , y_pred)
```

```python
2969/2969 #this is for majority class and accuracy is 100 percent
```

```python
from imblearn.over_sampling import BorderlineSMOTE
bpsmote = BorderlineSMOTE() 
```

```python
X_train_bpsmote , y_train_bpsmote = bpsmote.fit_resample(X_train.astype('float') , y_train)
```

```python
from collections import Counter 
print('Before BorederlineSMOTE :' , Counter(y_train))
print('After BorderlineSMOTE :' , Counter(y_train_bpsmote))
```

```python
model.fit(X_train_bpsmote , y_train_bpsmote) 
y_pred = model.predict(X_test) 
print(accuracy_score(y_test, y_pred))
pd.crosstab(y_test , y_pred)
```

```python
2902/2969 # the majority class accuracy becomes 97 percent after BorderlineSMOTE
```

```python
24/31 #the minority class accuracy becomes 77 percent after BorderlineSMOTE
```
